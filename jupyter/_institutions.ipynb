La Base de données locales référence plus de 63 000 guichets publics locaux (mairies, organismes sociaux, services de l'état, etc.). Elle fournit leurs coordonnées (adresses, téléphones, site internet, horaires d'ouverture, coordonnées de géolocalisation). Le schéma et sa documentation sont disponibles à l'adresse : Specifications-datagouv-base-de-donnees-locales-Service-public_V1.1.pdf

REMARQUE IMPORTANTE : Depuis le 4 juillet 2022, ce jeu de données a changé de format et de structure. Les données sont au format json.


```python
import pandas as pd
import numpy as np
base = '../data/premier_ministre/données_locales/all_latest/'
filename = '2022-12-12_002751-data.gouv_local.json'
data = pd.read_json(base + filename)
jsons = data.service.values
data = pd.DataFrame.from_dict(pd.json_normalize(jsons), orient='columns')
data.index.names = ['service']
data = data.applymap(lambda x: np.nan if x == '' or x == [] else x)
data = data.dropna(axis=1, how='all')
display(data)
```

On peut :
* supprimer les colonnes constantes :
    * `statut_de_diffusion` qui est toujours `true`
    * `version_type` qui est toujours `Publiable`
* passer la colonne `id` en index
* coder les deux valeurs possibles de `copyright` en DILA et MJ

NB :
* 35147 communes (codes INSEE) sont représentées
* cf. [Spec](https://echanges.dila.gouv.fr/OPENDATA/Base_donn%c3%a9es_locales/Specifications-datagouv-base-de-donnees-locales-Service-public_V1.1.pdf) SL vs. SIL dénotent "Service Local" vs. "Service Institutionnel Local"
* les types de service sont compris dans les noms, dont certains sont composés : [Type] - [Collectivité]

```python
dense_cols = [
    'categorie', 'copyright', 'date_creation',
    'nom', 'date_modification', 'id', 'code_insee_commune'
]

extract_1 = data[dense_cols]
extract_1 = extract_1.set_index('id')
extract_1.copyright = extract_1.copyright.apply(
    lambda x: 'MJ' if x == 'Ministère de la Justice' else 'DILA'
)
display(extract_1)
```

Exemple d'extraction de sous-ensemble : les mairies :

```python
extract_2 = extract_1[extract_1.nom.str.startswith('Mairie - ')]
display(extract_2)
```

Exemple 2 : seulement les mairies du Vaucluse

```python
extract_3 = extract_2[extract_2.code_insee_commune.str.startswith('84')]
display(extract_3)
```
