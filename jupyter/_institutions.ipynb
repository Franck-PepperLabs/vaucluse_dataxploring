# **Quelles sont les collectivités ?**

Nous nous intéressons ici à constituer un référentiel national le plus complet et actualisé des collectivités et autres organisations publiques et para-publiques.

Le point de départ de l'investigation excentrique est de lister toutes les questions du type, quelles sont :
* ma commune
* les communes limitrophes de ma commune
* la communauté de communes dont ma commune est membre
* les communautés de communes limitrophes
* le département et toutes les divisions territoriales intermédiaires dont est membre ma commune, ma communauté de commune, etc
* ...

Cette zone d'investigation est donc à croiser avec celle des élus qui permet de comprendre et rendre apparente la super-structure institutionnelle (administrative) et les incarnants des rôles et places qu'elle définit.

La Base de données locales référence plus de 63 000 guichets publics locaux (mairies, organismes sociaux, services de l'état, etc.). Elle fournit leurs coordonnées (adresses, téléphones, site internet, horaires d'ouverture, coordonnées de géolocalisation). En complément, sont indexés plus de 36 000 fichiers des communes, précisant la compétence géographique des guichets.
